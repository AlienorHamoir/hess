```python
import os
import eppy
from eppy import modeleditor
from eppy.modeleditor import IDF

# Change the working directory to ensure the correct path
os.chdir('/home/alienor/Documents/hess/resources/loads')
print(os.getcwd())  # Verify the new working directory

# Define the relative paths from the current working directory
iddfile = 'V1-4-0-Energy+.idd'
idffile = 'RefBldgSecondarySchoolNew2004_v1.4_7.2_3B_USA_CA_LOS_ANGELES.idf'

# Print the paths for debugging
print(f"IDD file path: {iddfile}")
print(f"IDF file path: {idffile}")

# Check if the files exist
if not os.path.isfile(iddfile):
    raise FileNotFoundError(f"IDD file not found: {iddfile}")
if not os.path.isfile(idffile):
    raise FileNotFoundError(f"IDF file not found: {idffile}")

# Set the IDD file
IDF.setiddname(iddfile)

# Read the IDF file
idf = IDF(idffile)

# Access the ZONE objects
zones = idf.idfobjects['ZONE']

# Write the zone names to a CSV file
with open('zones.csv', 'w') as f:
    f.write('ZoneName\n')
    for zone in zones:
        f.write(f"{zone.Name}\n")
```
